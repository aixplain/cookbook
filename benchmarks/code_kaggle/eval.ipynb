````python
import os
import json
import re

with open("data/code_kaggle_20240712.jsonl") as f:
    data = [json.loads(l) for l in f.readlines()]

for model in ["single_agent_llama", "orchestrator_agent_llama", "multi_agent_llama", "single_agent_gpt4"]:
    folder_path = f"models/{model}/results"

    y_real, y_pred = [], {}
    for idx, _ in enumerate(data):
        if data[idx]["image_expected"] is False:
            answers = data[idx]["gt_answers"]
            y_real.extend(answers)
            for i in range(3):
                if i not in y_pred:
                    y_pred[i] = []
                path = f"{folder_path}/{idx}_{i}.json"
                if os.path.exists(path):
                    with open(path) as f:
                        response = json.load(f)["output"]
                    try:
                        response = [w["answer"] for w in json.loads(response)]
                    except Exception:
                        a = re.findall("```json\n(.*)\n```", response, re.DOTALL)
                        if len(a) > 0:
                            response = [w["answer"] for w in json.loads(a[0])]
                        else: 
                            response = len(answers) * ["Z"]
                    if len(response) != len(answers):
                        response = len(answers) * ["Z"]
                    y_pred[i].extend(response)
                else:
                    y_pred[i].extend(len(answers) * ["Z"])
                

    from sklearn.metrics import accuracy_score
    for i in y_pred:
        print(f"Model {model} - Run {i} - Accuracy: {accuracy_score(y_real, y_pred[i])}")
````
