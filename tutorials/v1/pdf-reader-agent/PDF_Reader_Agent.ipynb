```python
# Get Docling model
docling = ModelFactory.get("677bee6c6eb56331f9192a91")
```

```python
# Create the index
index = IndexFactory.create(
    name="Car Insurance2",
    description="Index for technical documentation"
)

print(f"Index created! ID: {index.id}")
```

```python
# Process each PDF
records = []

for doc in data:
    print(f"Processing: {doc['title']}")

    # Extract text with Docling
    text = docling.run(doc["pdf_url"]).data

    # Create a record with metadata
    record = Record(
        id=doc["id"],
        value=text,
        attributes={
            "author": doc["author"],
            "title": doc["title"],
            "source": doc["pdf_url"]
        }
    )
    records.append(record)

print(f"Extracted {len(records)} documents")
```

```python
# Upload to index in batches
for i in range(0, len(records), 500):
    batch = records[i:i + 500]
    index.upsert(batch)
    print(f"Uploaded batch {i//500 + 1}")

# Verify upload
print(f"Total documents in index: {index.count()}")
```

```python
# Define agent instructions
agent_instructions = '''
You are an AI assistant specialized in retrieving and providing
information from PDF documents.

IMPORTANT RULES:
1. ONLY use your search tool to answer questions
2. DO NOT use your internal knowledge
3. ALWAYS cite the source document and author

PROCESS:
1. Use the search tool to query the PDF index
2. Extract relevant content from the results
3. Provide a clear answer with document URL from metadata
4. If you can't find the answer in the PDFs, say so clearly
'''

# Create the agent
agent = AgentFactory.create(
    name="Technical PDF Agent",
    description="An agent that answers questions using PDF documents",
    instructions=agent_instructions,
    tools=[
        AgentFactory.create_model_tool(model=index.id)
    ]
)

print(f"Agent created! ID: {agent.id}")
```

```python
# First query
response = agent.run(
    "What roadside cover do I have?"
)

print(response.data.output)
```

```python
# Get session ID for follow-up questions
session_id = response.data.session_id

# Ask a follow-up question
response = agent.run(
    "Who would I contact for roadside and home rescue?",
    session_id=session_id
)

print(response.data.output)
```

```python
# Another follow-up
response = agent.run(
    "Can you summarize the main points?",
    session_id=session_id
)

print(response.data.output)
```

```python
# Deploy the agent
agent.deploy()

print(f"Agent deployed! Access it at:")
print(f"https://platform.aixplain.com/discover/agent/{agent.id}")
```
