# What is the Video Gen Agent?

The goal of this agent is to make professional short-video production conversational. You describe the video you want — the scene, mood, movement, style, and format — and the agent generates it using BytePlus Seedance.

## Agent Architecture

The agent uses one tool:

- **BytePlus Seedance** — a production-grade text-to-video and image-to-video model that generates 2–12 second clips at up to 1080p in any standard aspect ratio (16:9, 9:16, 1:1, 4:3, 21:9). Accessible directly from the aiXplain platform — no separate API key required.

## Agent Workflow

1. **Understand the request** — extract the scene, style, format, and any reference image from the user's message.
2. **Compose the prompt** — translate the request into a precise, Seedance-optimised video prompt.
3. **Generate** — call the Seedance model with the prompt, resolution, aspect ratio, and duration.
4. **Deliver** — return the video URL when generation completes.

```python
# Load the BytePlus Seedance video generation model from the aiXplain platform
SEEDANCE_MODEL_PATH = "bytedance/seedance-1-5-pro/bytedance"

video_tool = aix.Model.get(SEEDANCE_MODEL_PATH)
print(f"Loaded: {video_tool.name}")
```

# Build Your Agent

To create an agent, define:

* A unique **name** and **description** for its purpose.
* The **tool** it will use — the BytePlus Seedance model loaded from the aiXplain platform.
* The **instructions** that guide the agent's prompt engineering and output quality.

```python
video_agent = aix.Agent(
    name="Video Gen Agent",
    description="Generates short videos from text prompts or reference images using BytePlus Seedance. Produces 2–12 second clips for ads, social media, brand content, and explainers.",
    instructions="""
    You are a professional video production assistant powered by the BytePlus Seedance video generation model.

    When a user asks for a video:
    1. Extract the key elements from their request:
       - Subject and action (what is happening)
       - Visual style (cinematic, animated, realistic, product shot, etc.)
       - Aspect ratio: 16:9 (landscape/YouTube), 9:16 (vertical/TikTok/Reels), 1:1 (square/Instagram)
       - Duration: 2–12 seconds (default to 5s if not specified)
       - Resolution: 480p, 720p, or 1080p (default to 720p)

    2. Write a precise, descriptive video prompt:
       - Lead with the main subject and action
       - Include camera movement (slow zoom, pan, static, tracking shot)
       - Describe lighting, mood, and visual atmosphere
       - Keep prompts under 200 words — clarity beats length
       - Do NOT add details the user didn't mention (duration, aspect ratio specs, style labels)
         unless they are clearly implied or requested

    3. Call the Seedance model with the composed prompt.

    4. Return the video URL when generation is complete.
       Provide the full URL as plain text so it is always clickable.

    Supported use cases:
    - Product advertisements and demos
    - Social media clips (TikTok, Instagram Reels, YouTube Shorts)
    - Brand videos and intros
    - Explainer and concept visuals
    - Scene and mood B-roll

    If the user provides a reference image, include it as an image input to the model
    so Seedance uses it as the first frame or visual reference.
    """,
    tools=[video_tool],
)
video_agent.save()
print(f"Agent created: {video_agent.id}")
```

# Run your Agent

To ensure your agent is performing as expected, run it using the `run` method with sample inputs. Analyze the outputs, verify their accuracy, and debug any issues by inspecting the agent's configurations, tools, and instructions.

> **Note:** Video generation is asynchronous and typically takes 1–3 minutes. The agent will return the video URL once generation is complete.

```python
# Text-to-video: product ad
result = video_agent.run(
    """Generate a 5-second 9:16 product ad for a sleek black coffee machine on a marble countertop, 
    cinematic lighting, slow zoom in""", max_iterations=20
)
print(result.data.output)
```

```python
# Social media clip
result2 = video_agent.run(
    """Create a short vertical video of a sunset over the ocean with waves, 
    warm golden tones, for Instagram Reels""", max_iterations=20
)
print(result2.data.output)
```

```python
# Multi-turn: refine the video
result3 = video_agent.run(
    "Make it more dramatic — stormy sky, bigger waves, darker mood",
    session_id=result2.data.session_id
)
print(result3.data.output)
```
