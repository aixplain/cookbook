# What is the Web Search Agent?

The goal of this agent is to give an AI assistant accurate, up-to-date information from the live web. Tavily is designed specifically for AI agents — it returns clean, structured results without the noise of a traditional search engine.

## Agent Architecture

The agent uses one tool:

- **Tavily Web Search** — an AI-optimised search tool available directly on the aiXplain platform. It supports general research, deep-research mode for complex questions, and news-focused queries. No separate Tavily API key is required.

## Agent Workflow

1. **Understand the request** — determine the query, topic type (general or news), and required depth.
2. **Search** — call the Tavily tool with the appropriate parameters.
3. **Synthesise** — summarise findings, cite sources, and present a clear answer.

```python
# Load the Tavily Web Search tool from the aiXplain platform
search_tool = aix.Tool.get("tavily/tavily-web-search/tavily")
print(f"Loaded: {search_tool.name}")
```

# Build Your Agent

To create an agent, define:

* A unique **name** and **description** for its purpose.
* The **tool** it will use — the Tavily Web Search tool loaded from the aiXplain platform.
* The **instructions** that guide the agent's search behaviour and result presentation.

```python
search_agent = aix.Agent(
    name="Web Search Agent",
    description="Research assistant that searches the live web using Tavily's AI-optimised search API to answer questions with up-to-date, sourced information.",
    instructions="""
    You are a research assistant with access to the Tavily web search tool.

    When a user asks a question:
    1. Identify whether it needs general research, deep research, or news.
    2. Call the Tavily Web Search tool with a well-formed query.
       - For simple factual questions: standard search.
       - For complex research questions: use deep search mode.
       - For recent events or breaking news: use news topic mode.
    3. Summarise the results clearly:
       - Lead with the direct answer.
       - Support with key details from the search results.
       - Always cite sources with titles and URLs.
    4. If the first search doesn't fully answer the question, refine the query and search again.

    Guidelines:
    - Never present information as fact without grounding it in a search result.
    - Flag when information may be outdated or rapidly changing.
    - For multi-part questions, run separate searches for each part.
    - Keep answers concise — surface the most relevant information, not everything returned.
    """,
    tools=[search_tool],
)
search_agent.save()
print(f"Agent created: {search_agent.id}")
```

# Run your Agent

To ensure your agent is performing as expected, run it using the `run` method with sample inputs. Analyze the outputs, verify their accuracy, and debug any issues by inspecting the agent's configurations, tools, and instructions.

```python
# General research question
result = search_agent.run("What are the latest developments in AI agents?")
print(result.data.output)
```

```python
# News query
result2 = search_agent.run("What is the latest news about OpenAI this week?")
print(result2.data.output)
```

```python
# Deep research
result3 = search_agent.run(
    "Give me a comprehensive overview of how retrieval-augmented generation works"
)
print(result3.data.output)
```

```python
# Multi-turn: follow up on previous result
result4 = search_agent.run(
    "What are the main limitations of RAG systems?",
    session_id=result3.data.session_id
)
print(result4.data.output)
```
